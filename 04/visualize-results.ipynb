```python
data = {}
with open('out.txt', 'r') as fin:
    for line in fin:
        tokens = line.split()
        key, values = tokens[0], [int(x) for x in tokens[1:]]
        data[key] = values
```

```python
name = list(data.keys())[4]
f, ax = plt.subplots(1)
ax.plot(linspace(0, 1, 100), data[name])
ax.set_ylim(ymin=0)
ax.set_title(name)
```

```python
f = plt.figure(figsize=(10, 30))
for i in range(10):
    name = list(data.keys())[i]
    ax = f.add_subplot(5, 2, i + 1)
    ax.plot(linspace(0, 1, 100), data[name])
    ax.set_ylim(ymin=0)
    ax.set_title(name)
```
